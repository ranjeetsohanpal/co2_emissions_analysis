```python
import pandas as pd
```

```python
co2_filepath = '../CO2_emissions_data.csv'
co2_df = pd.read_csv(co2_filepath)
```

```python
co2_df.head()
```

```python
countries_filepath = 'countries_by_continent.csv'
countries_df = pd.read_csv(countries_filepath)
```

```python
countries_df.head()
```

1. Clean countries_df : Remove Unnamed:0 column
2. Clean co2_df : filter out the countries and join with the countries_df to get another column of continent for every country.

```python
countries_df.columns
```

```python
countries_df = countries_df.iloc[:,1:]
```

```python
countries_df.head()
```

```python
countries_df.info()
```

```python
co2_df['Description'].unique()
```

```python
co2_df.shape
```

```python
co2_df_countries = co2_df[co2_df['Description'] == 'Country'] # Keep only countries in the data
```

```python
co2_df_countries.shape
```

```python
co2_df_countries.columns[:5]
```

```python
co2_df_countries.rename(columns={'Name' : 'Country'},inplace=True)
```

```python
co2_df_countries.info()
```

```python
co2_df_countries = co2_df_countries.iloc[:,1:] #removing the first column
```

```python
co2_df_countries.head()
```

```python
# joining the dataframe
co2_df_countries = co2_df_countries.merge(
    countries_df[['Country', 'Continent']],  
    on='Country',
    how='left'
)
```

```python
co2_df_countries.tail(10)
```

```python
co2_df_countries[co2_df_countries['year'] == 2023 ].info()
```

```python
# A lot of important KPIs such as gdp, energy_per_gdp, consumption_co2, co2_per_gdp. So 2023 would be removed from the data

co2_df_countries = co2_df_countries[co2_df_countries['year'] != 2023]
```

```python
co2_df_countries.tail(10)
```

```python
co2_df_countries.to_csv('co2_emissions_cleaned.csv')
```
